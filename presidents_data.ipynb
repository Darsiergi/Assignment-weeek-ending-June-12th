```python
import networkx as nx
```

```python
president_graph = nx.from_pandas_edgelist(us_presidents, source="From", target="To")
```

# # Analysis

- Diameter
- Degree of Connection
- Most Influential
- Most important connection
- Best betweener
- Shortest path

```python
# Diameter
nx.diameter(president_graph)
```

```python
# Degree of Connection in General
nx.degree(president_graph)
```

```python
# Degree of Connection Specific Instance
nx.degree(president_graph, "Bush")
```

```python
# Most Influential Presidents
nx.degree_centrality(president_graph)
```

```python
# Sorted Equivalent as above
sorted(nx.degree_centrality(president_graph).values())
most_influential = nx.degree_centrality(president_graph)
```

```python
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

# # Visualization

### Two Ways of Rendering Graphs
- nx.draw
- nx.draw_networkx

```python
# Loading relevant package
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# First Way
nx.draw(president_graph)
```

```python
nx.draw(president_graph, with_labels=True)
```

```python
nx.draw(president_graph, with_labels=True, node_color='y')
```

```python
# Second Way
nx.draw_networkx(president_graph)
```

```python
plt.figure(figsize=(10, 10))
nx.draw_networkx(president_graph)
plt.show()
```
